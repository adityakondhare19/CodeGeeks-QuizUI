```
# ============================================================
# FINAL FIXED MEDICAL EXPERT SYSTEM (WORKING VERSION)
# ============================================================

disease_db = {
    "Common Cold": {
        "must": ["runny nose", "sneezing"],
        "extra": ["mild fever", "sore throat", "cough", "tiredness"]
    },
    "Flu": {
        "must": ["high fever", "body pain"],
        "extra": ["fatigue", "cough", "sore throat", "headache", "chills"]
    },
    "COVID-19": {
        "must": ["fever"],
        "extra": ["dry cough", "fatigue", "loss of taste", "loss of smell",
                  "breathing problem", "headache"]
    },
    "Malaria": {
        "must": ["high fever", "chills"],
        "extra": ["sweating", "headache", "vomiting", "body pain"]
    },
    "Typhoid": {
        "must": ["high fever"],   # FIXED (was long fever)
        "extra": ["stomach pain", "headache", "weakness", "loss of appetite"]
    }
}

# Generate symptoms list
symptom_list = sorted(set(
    s for d in disease_db.values()
    for s in d["must"] + d["extra"]
))

# ---------------- DIAGNOSIS ----------------
def find_disease(user_symptoms):
    results = []

    for disease, rule in disease_db.items():
        must = rule["must"]
        extra = rule["extra"]

        # Count matches instead of strict check
        must_match = sum(1 for s in must if s in user_symptoms)
        extra_match = sum(1 for s in extra if s in user_symptoms)

        total = len(must) + len(extra)
        matched = must_match + extra_match

        if matched == 0:
            continue

        confidence = round((matched / total) * 100, 1)

        results.append((disease, confidence))

    results.sort(key=lambda x: x[1], reverse=True)
    return results

# ---------------- MAIN ----------------
if __name__ == "__main__":
    print("\n===== Medical Expert System =====")

    while True:
        print("\nSymptoms:")
        for i, s in enumerate(symptom_list, 1):
            print(f"{i}. {s}")

        raw = input("\nEnter numbers (e.g. 1,2): ").strip()

        try:
            indices = [int(x)-1 for x in raw.split(',')]
            user_symptoms = [symptom_list[i] for i in indices if 0 <= i < len(symptom_list)]
        except:
            print("❌ Invalid input")
            continue

        print("\nSelected:", user_symptoms)

        result = find_disease(user_symptoms)

        if not result:
            print("No disease matched.")
        else:
            print("\nPossible Diseases:")
            for d, c in result:
                print(f"{d} → {c}%")

            print(f"\nMost likely: {result[0][0]}")

        if input("\nAgain? (y/n): ").lower() != 'y':
            break
```
