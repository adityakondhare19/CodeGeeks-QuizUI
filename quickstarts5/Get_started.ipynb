```
# ============================================================
# AI Assignment 5: Simple Chatbot (Modified Version)
# ============================================================

import re
import random
import datetime

# Patterns
response_patterns = [
    (r'\b(hi|hello|hey)\b',
     ["Hello! How can I help you?",
      "Hi! What do you need?",
      "Hey there!"]),

    (r'\bhow are you\b',
     ["I am fine! What about you?",
      "Doing great! How are you?"]),

    (r'\bwhat is your name\b|\bwho are you\b',
     ["I am SmartBot, your assistant.",
      "You can call me SmartBot!"]),

    (r'\btime\b',
     [f"Current time is {datetime.datetime.now().strftime('%I:%M %p')}"]),

    (r'\bdate\b|\bday\b',
     [f"Today is {datetime.datetime.now().strftime('%A, %d %B %Y')}"]),

    (r'\bai\b|\bartificial intelligence\b',
     ["AI means machines can think like humans.",
      "Artificial Intelligence helps machines learn."]),

    (r'\bjoke\b',
     ["Why did the computer sleep? Because it had too many tabs open 😄",
      "Why do coders hate nature? Too many bugs!"]),

    (r'\bthank(s| you)\b',
     ["Welcome!", "No problem!", "Glad to help!"]),

    (r'\b(bye|exit|quit)\b',
     ["Bye! Take care!",
      "Goodbye!",
      "See you soon!"]),

    (r'.*',
     ["I did not understand. Try again.",
      "Can you rephrase that?",
      "Ask me about time, jokes, or math."])
]

# -------- SIMPLE MATH --------
def calculate(text):
    match = re.search(r'(\d+)\s*([\+\-\*\/])\s*(\d+)', text)
    if match:
        x = int(match.group(1))
        op = match.group(2)
        y = int(match.group(3))

        if op == '+': return f"{x+y}"
        if op == '-': return f"{x-y}"
        if op == '*': return f"{x*y}"
        if op == '/' and y != 0: return f"{x/y:.2f}"
        if op == '/' and y == 0: return "Cannot divide by zero!"
    return None

# -------- MEMORY --------
name_memory = ["friend"]

# -------- RESPONSE FUNCTION --------
def chatbot_reply(user_text):
    text = user_text.lower().strip()

    # Save name
    name_match = re.search(r'my name is (\w+)', text)
    if name_match:
        name_memory[0] = name_match.group(1).capitalize()
        return f"Nice to meet you {name_memory[0]}!"

    # Recall name
    if re.search(r'what is my name|remember my name', text):
        return f"Your name is {name_memory[0]}"

    # Math
    math_ans = calculate(text)
    if math_ans:
        return f"Answer: {math_ans}"

    # Pattern matching
    for pattern, replies in response_patterns:
        if re.search(pattern, text):
            return random.choice(replies)

    return "I don't know how to respond."

# ---------------- MAIN ----------------
if __name__ == "__main__":
    print("=" * 40)
    print("        Welcome to SmartBot")
    print("   Type 'bye' to exit")
    print("=" * 40)

    while True:
        try:
            user_text = input("\nYou : ").strip()
            if not user_text:
                continue

            bot_reply = chatbot_reply(user_text)
            print("Bot :", bot_reply)

            if re.search(r'\b(bye|exit|quit)\b', user_text.lower()):
                break

        except KeyboardInterrupt:
            print("\nBot : Goodbye!")
            break
```
